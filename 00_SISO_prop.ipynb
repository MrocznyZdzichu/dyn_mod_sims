```python
import math

import Objects
import numpy_utils
import Objects_Plotter as sim_plt


gain_matrix = .4
SISO_charger = Objects.Proportional_object(gain=gain_matrix, sampling_time=0.1)
```

```python
u = []
for tt in range(0, 100):
    u.append(math.tanh(0.03*tt))
```

```python
SISO_charger.reset_history()
for ut in u:
    SISO_charger.simulate_step(ut)
```

```python
sim_plt.plot_object_history(SISO_charger)
```
