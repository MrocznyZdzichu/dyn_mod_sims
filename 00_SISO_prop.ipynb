```python
import math
import matplotlib.pyplot as plt

import Objects
import numpy_utils


gain_matrix = .4
SISO_charger = Objects.Proportional_object(gain=gain_matrix, sampling_time=0.1)
```

```python
u = []
for tt in range(0, 100):
    u.append(math.tanh(0.03*tt))
```

```python
SISO_charger.reset_history()
for ut in u:
    SISO_charger.simulate_step(ut)

u = [float(ut[0][0]) for ut in SISO_charger.get_input_hist()]
t = SISO_charger.get_timestamps()
y = [float(yt[0][0]) for yt in SISO_charger.get_output_hist()]
```

```python
plt.plot(t, u, label='Input signal')
plt.plot(t, y, label='Output signal')
plt.title('SISO Proportional object simulation')
plt.xlabel('Time [s]')
plt.ylabel('Signal value')
plt.grid()
sink = plt.legend()
```
