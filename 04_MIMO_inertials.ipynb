```python
from numpy_utils import converters
from scipy.signal import TransferFunction
import numpy as np
import scipy.signal
import matplotlib.pyplot as plt
```

```python
tf1 = TransferFunction([1], [1, 1])
tf2 = TransferFunction([1], [1, 1.5, 1])

tf_matrix = [[tf1, 0], [0, tf2], [tf1, tf2]]

tf_matrix_arr = converters.to_array_converter().convert_any_type(tf_matrix)

from Objects import TF_object
obj = TF_object(sampling_time=0.1, tf_matrix=tf_matrix_arr)
```

```python
conv = converters.to_vector_converter()
for i in range(0, 100):
    u1=0 if i < 20 else 1
    u2=0 if i < 50 else 1
    u = conv.convert_any_type([u1, u2])
    obj.simulate_step(u)
```

```python
t = obj.get_timestamps()

u1 = [u[0][0] for u in obj.get_input_hist()]
u2 = [u[1][0] for u in obj.get_input_hist()]

y1 = [y[0][0] for y in obj.get_output_hist()]
y2 = [y[1][0] for y in obj.get_output_hist()]
y3 = [y[2][0] for y in obj.get_output_hist()]
```

```python
fig, axes = plt.subplots(3, 1, figsize=(16, 24))
ax1, ax2, ax3 = axes

ax1.plot(t, u1, '--', label='Input signal 1')
ax1.plot(t, u2, '--', label='Input signal 2')
ax1.plot(t, y1, label='Output signal 1')
ax1.set_title('System response 1')
ax1.set_xlabel('Time [s]')
ax1.set_ylabel('Signal value')
sink = ax1.legend()

ax2.plot(t, u1, '--', label='Input signal 1')
ax2.plot(t, u2, '--', label='Input signal 2')
ax2.plot(t, y2, label='Output signal 2')
ax2.set_title('System response 2')
ax2.set_xlabel('Time [s]')
ax2.set_ylabel('Signal value')
sink = ax2.legend()

ax3.plot(t, u1, '--', label='Input signal 1')
ax3.plot(t, u2, '--', label='Input signal 2')
ax3.plot(t, y3, label='Output signal 3')
ax3.set_title('System response 3')
ax3.set_xlabel('Time [s]')
ax3.set_ylabel('Signal value')
sink = ax3.legend()
```
