## MIMO complex example

```python
time_consts = [
    [[0, 0, 1], [1, 0, 0], [0, 1, 0]]
    ,[[0, 0, 0.5], [0, 0, 0.1], [0, 0.01, 0.1]]
    ,[[0.2, 0, 0], [3, 0, 1], [0.7, 0, 0]]
]
gains = [
    [.002, 0.001, 0.003],
    [.02, 0, 0],
    [.0, 0.003, 0.001]
]

obj = Integral_Object(time_consts, gains, sampling_time=0.1)
```

```python
for i in range(0, 300):
    u1 = 0 if i < 30 or i > 200 else 1
    u2 = 0 if i < 50 or i > 180 else 1
    u3 = 0 if i < 80 or i > 220 else 0.6
    u = cnvs.to_vector_converter().convert_any_type([u1, u2, u3])
    obj.simulate_step(u)

sim_plt.plot_object_history(obj, figsize=(24, 30))
```
