## Stability for T0 > 0 
T1 > 0 - damping of oscilations \
T1 = 0 - no damping, persistent oscilations \
T1 < 0 - strenghtening of oscilations - losage of stability

```python
gains = [
    [1],
    [1],
    [1]
]
time_consts = [
    [[.2, 0.2]],
    [[.2, 0]],
    [[.2, -0.2]]
]
obj = Oscilating_Object(time_consts, gains, sampling_time=0.1)
```

```python
for i in range(0, 200):
    u = 0 if i < 20 or i > 80 else 1
    obj.simulate_step(u)

sim_plt.plot_object_history(obj, figsize=(24, 20))
```
